```python
# library
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Create a dataset
df = pd.DataFrame(np.random.random((5,5)), columns=["a","b","c","d","e"])

# show value in heatmap of float format # fmt="f" means show value in float format
# cmap="YlGnBu" means show color in yellow-green-blue
# annot=True means show value in heatmap
p2 = sns.heatmap(df, annot=True, fmt="f", cmap="YlGnBu")
# show title
p2.set_title('Heatmap')

plt.show()


```

```python
#import library chart, numpy, pandas.....
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Crearte Data 
height = [3, 12, 5, 18, 45]
bars = ('ĐHTT', 'THA', 'ĐTH', 'CNTT', 'SCU')
# Tạo ra các số trong trục x 
#x_pos = np.arange(len(bars))

# Create a Figure and an Axes # Initialize layout
fig, ax = plt.subplots()
plt.figsize = (12, 12)

#Tạo chart
bh = ax.bar(bars,height);

#change color all bar
bh[0].set_color('r')    # ĐHTT  - Red   
bh[1].set_color('g')    # THA   - Green
bh[2].set_color('b')    # ĐTH   - Blue
bh[3].set_color('y')    # CNTT  - Yellow
bh[4].set_color('c')    # SCU   - Cyan


# Thêm tiêu đề
ax.set_title('Tiêu đề Chart Demo')
# Show giá trị trên mỗi cột
ax.bar_label(bh, padding=1)


# Show chart
plt.show()
```

```python
# library
import matplotlib.pyplot as plt

# create data: an array of values
size_of_groups = [12, 11, 3, 30,55]

# Create a Figure and an Axes # Initialize layout
fig, ax = plt.subplots()

# show value in pie chart # autopct='%1.1f%%' means show value in percentage
ax.pie(size_of_groups, autopct='%1.1f%%')
# show legend
ax.legend(['A', 'B', 'C', 'D','E'])
# show title
ax.set_title('Pie Chart')

plt.show()
```

```python
# library
import matplotlib.pyplot as plt
import numpy as np

# create data
x = np.arange(0, 10, 0.1)
y = np.sin(x)

# Create a Figure and an Axes # Initialize layout
fig, ax = plt.subplots()

# show line chart
ax.plot(x, y)
# show title
ax.set_title('Line Chart')

plt.show()
```

```python
#import library chart, numpy, pandas, seaborn.....
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Create a dataset
df = pd.DataFrame(np.random.random((5,5)), columns=["a","b","c","d","e"])

# Create a Figure and an Axes # Initialize layout
fig, ax = plt.subplots()

# show area chart
ax.stackplot(df.index, df["a"], df["b"], df["c"], df["d"], df["e"])
# show title
ax.set_title('Area Chart')

plt.show()
```

```python
#import library chart, numpy, pandas, seaborn.....
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a dataset multi bar chart
df = pd.DataFrame(np.random.random((5,5)), columns=["a","b","c","d","e"])

# Create a Figure and an Axes # Initialize layout
fig, ax = plt.subplots()

# show multi bar chart

ax.bar(df.index, df["a"], label="a")
ax.bar(df.index, df["b"], bottom=df["a"], label="b")
ax.bar(df.index, df["c"], bottom=df["a"]+df["b"], label="c")
ax.bar(df.index, df["d"], bottom=df["a"]+df["b"]+df["c"], label="d")
ax.bar(df.index, df["e"], bottom=df["a"]+df["b"]+df["c"]+df["d"], label="e")

# show legend
ax.legend()
# show title
ax.set_title('Multi Bar Chart')

plt.show()
```
